```python
class student:
  def __init__(self, stu_id, name, marks, dept):
    self.std_id = std_id
    self.name = name
    self.marks = marks
    self.dept = dept

  def result(self):
    if self.marks >= 80:
      print("Student pass with grade A")

    elif self.marks >= 65:
      print("Student pass with grade B")

    elif self.marks >= 50:
      print("Student pass with grade C")

    elif self.marks >= 35:
      print("Student pass with grade D")

    else:
      print("Student Faill (marks <35)")

  def info(self):
    return f"Student ID: {self.std_id} | Student Name: {self.name} | Student Marks: {self.marks} | Department Name: {self.dept}"

students = []
while True:
  print(f"\n 1. Add Student\n 2. Show All Student\n 3. Search Student\n 4. Exit")
  choice = input("Enter your choice: ")


  if choice == "1":
    std_id = input("Enter Student ID: ")
    name = input("Enter student name: ")
    marks = float(input("Enter your marks: "))
    dept = input("Enter Department name: ")
    studt = student(std_id, name, marks, dept)
    print(f"Student {name} Added Successfully")
    students.append(studt)

  elif choice == "2":
    for studt in students:
      print(studt.info())

  elif choice == "3":
    std_id = input("Enter Student id to search: ")
    found = False
    for studt in students:
      if studt.std_id.lower() == std_id.lower():
        print(studt.info())
        found = True
      if not found:
        print(f"no Student found in {std_id} number")


  elif choice == "4":
    break

  else:
    print("Invalid choice")




```
