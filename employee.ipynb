```python
class employee:
  def __init__(self, emp_id, name, salary, dept):
    self.emp_id = emp_id
    self.name = name
    self.salary = salary
    self.dept = dept

  def apply_bonus(self):
    if self.salary<30000:
      bonus = self.salary * 0.20

    elif self.salary < 60000:
      bonus = self.salary * 0.10

    else:
      bonus = self.salary * 0.05

    self.salary += bonus
    return bonus

  def info(self):
    return f"Employee ID: {self.emp_id} | Employee Name: {self.name} | Salary: {self.salary} | Deptment: {self.dept}"

employees = []
while True:
  print(f"\n 1. Add Employee\n 2. Apply Bonus\n 3. Show All \n 4. Search by dept\n 5. Exit")
  choice = input("Enter your choice: ")


  if choice == "1":
    emp_id = input("Enter Employee ID: ")
    name = input("Enter Employee Name: ")
    salary = float(input("Enter Employee Salary: "))
    dept = input("Enter Employee Dept: ")
    emp = employee(emp_id, name, salary, dept)
    print(f"Employee {name} Added Successfully")
    employees.append(emp)

  elif choice =="2":
    for emp in employees:
      bonus = emp.apply_bonus()
      print(f"bonus of {bonus} applied to {emp.name}")

  elif choice == "3":
    for emp in employees:
      print(emp.info())

  elif choice == "4":
    dept = input("Enter dept to search:")
    found = False
    for emp in employees:
      if emp.dept.lower() == dept.lower():
        print(emp.info())
        found = True
      if not found:
        print(f"No employee found in {dept} department")

  elif choice == "5":
    break
  else:
    print("Invalid Choice")

```
